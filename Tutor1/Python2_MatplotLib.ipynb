```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
data = pd.read_csv("Tutorial-1-tips.csv")
```

```python
### Problem 1
sb.scatterplot(x = "total_bill", y = "tip", hue = "time", data = data)
```

```python
### Problem 2
days = data["day"].value_counts()
plt.pie(days, labels = days.index)
```

```python
### Problem 3
sb.barplot(x = "sex", y = "total_bill", data = data)
```

```python
### Problem 4
sb.distplot(data["total_bill"])
```

```python
### Problem 5
sb.jointplot(data["total_bill"], data["tip"])
```

```python
### Problem 6
sb.pairplot(data, hue = "time")
```

```python
### Problem 7
sb.stripplot(x = "day", y = "total_bill", data = data)
```
